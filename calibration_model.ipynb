```python
from datetime import datetime, timedelta
from itertools import combinations
import random


# Function to create a combined appliance dictionary with pretend data for combined use
def change_in_points(appliance_dict):
    combined_data = {}

    # Iterate over each key-value pair in the appliance_dict
    for appliance, data in appliance_dict.items():
        # Extract timestamps and water meter readings
        timestamps = [entry[0] for entry in data]
        readings = [int(entry[1])/10.0 for entry in data]

        # Calculate change in water meter reading
        changes_readings = [0] + [readings[i+1] - readings[i] for i in range(len(readings)-1)]

        # Calculate change in timestamps
        changes_timestamps = [0] + [(datetime.strptime(timestamps[i+1], '%m/%d/%Y %H:%M:%S') - datetime.strptime(timestamps[i], '%m/%d/%Y %H:%M:%S')).total_seconds() for i in range(len(timestamps)-1)]

        # Round down changes in timestamps to the nearest 10 seconds
        rounded_changes_timestamps = [10 if change == 11 else change for change in changes_timestamps]

        # Combine timestamps and changes
        combined_data[appliance] = list(zip(rounded_changes_timestamps, changes_readings))

    return combined_data

def appliance_2(appliance_dict):

  # create new appliance key
  new_appliance = ''
  for key in appliance_dict.keys():
    new_appliance += key + " + "


  new_dic = {}

  # Find the longer list
  longest_list_key = max(appliance_dict, key=lambda k: len(appliance_dict[k]))
  longest_list = appliance_dict[longest_list_key]

  # Find the shorter list
  shorter_list_key = min(appliance_dict, key=lambda k: len(appliance_dict[k]))
  shorter_list = appliance_dict[shorter_list_key][1:]

  new_dic[new_appliance] = longest_list[:]

  c = True
  while c:
    # Select a random index to insert the shorter list
    index_to_insert = random.randint(0, len(longest_list) - len(shorter_list))

    # Get the difference in the first timestamp between the two lists

    time_diff = longest_list[index_to_insert][0] - shorter_list[0][0]

    if time_diff == 0:
      c = False
      for i in range(len(shorter_list)):
        new_dic[new_appliance][index_to_insert + i] = (
                shorter_list[i][0],
                shorter_list[i][1] + longest_list[index_to_insert + i][1],
            )

  return new_appliance, new_dic[new_appliance]

def generate_points(dictionary):
    all_combinations = []
    keys = list(dictionary.keys())

    # Explore combinations of 2 or more items
    for r in range(2, len(keys) + 1):
        for combo in combinations(keys, r):
            combination_dict = {key: dictionary[key] for key in combo}
            all_combinations.append(combination_dict)

    combined_data = {}

    for combo in all_combinations:

      if len(combo) == 2:
        longest_list_key, longest_list = appliance_2(combo)
        dictionary[longest_list_key] = longest_list

      else:
        longest_list_key = max(combo, key=lambda k: len(combo[k]))
        longest_list = combo[longest_list_key]

        # Find the shorter lists
        shorter_lists_keys = [k for k in combo if k != longest_list_key]
        shorter_lists = [combo[k] for k in shorter_lists_keys]

        for k, shorter_list in enumerate(shorter_lists):
          combo = {longest_list_key: longest_list, shorter_lists_keys[k]: shorter_list}
          longest_list_key, longest_list = appliance_2(combo)

        dictionary[longest_list_key] = longest_list

    return dictionary

def undo_change_in_points(combined_data):
    # Create a dictionary to store the original data
    original_data = {}

    # Iterate over each key-value pair in the combined_data
    for appliance, data in combined_data.items():
        # Initialize lists to store timestamps and actual water meter readings
        timestamps = []
        readings = []

        # Iterate over each data point
        for i, (timestamp, change) in enumerate(data):
            # Calculate the actual water meter reading
            if i == 0:
                actual_time = timestamp
                actual_reading = change
            else:
                actual_time += timestamp
                actual_reading += change

            # Append the timestamp and actual reading to the lists

            dt_object = datetime.utcfromtimestamp(actual_time)

            a = dt_object.strftime('%m/%d/%Y %H:%M:%S')

            timestamps.append(a)
            readings.append(actual_reading)

        # Store the original data in the dictionary
        original_data[appliance] = list(zip(timestamps, readings))

    return original_data


r = change_in_points(appliance_dict)
a = r.copy()
d = generate_points(r)
calibration_data_dic = undo_change_in_points(d)
```

```python
import plotly.graph_objs as go
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import plotly.graph_objs as go
from scipy.optimize import curve_fit
import numpy as np


# Function to extract features from water meter readings
def extract_features(d):
  final = []
  for data in d:
    # Convert to DataFrame
    df = pd.DataFrame(data, columns=['timestamp', 'reading'])
    df['timestamp'] = pd.to_datetime(df['timestamp'], format='%m/%d/%Y %H:%M:%S')
    df['reading'] = df['reading'].astype(int)/10

    # Calculate duration in minutes
    duration = (df['timestamp'].iloc[-1] - df['timestamp'].iloc[0]).total_seconds()/60

    # Calculate total water consumption
    total_water_consumption = df['reading'].iloc[-1] - df['reading'].iloc[0]

    # Calculate flow rates
    df['flow_rate'] = df['reading'].diff() / (df['timestamp'].diff().dt.total_seconds()/60)


    # Calculate mean, max, min flow rates
    mean_flow_rate = df['flow_rate'].mean()
    max_flow_rate = df['flow_rate'].max()
    min_flow_rate = df['flow_rate'].min()

    # Calculate the number of flow rate changes
    flow_rate_changes = df['flow_rate'].ne(df['flow_rate'].shift()).sum()

    # Calculate the number of spikes (defined as significant jumps in flow rate)
    spike_threshold = 10  # You can adjust the spike threshold as needed
    number_of_spikes = (df['flow_rate'].diff().abs() > spike_threshold).sum()

    # Calculate the duration of high flow (flow rate > certain threshold)
    high_flow_threshold = 5  # You can adjust the high flow threshold as needed
    high_flow_periods = df['flow_rate'] > high_flow_threshold
    duration_of_high_flow = high_flow_periods.sum() * df['timestamp'].diff().dt.total_seconds().mean()

    # Return extracted features
    features = {
        'duration': duration,
        'total_water_consumption': total_water_consumption,
        'mean_flow_rate': mean_flow_rate,
        'max_flow_rate': max_flow_rate,
        'min_flow_rate': min_flow_rate,
        'flow_rate_changes': flow_rate_changes,
        'number_of_spikes': number_of_spikes,
        'duration_of_high_flow': duration_of_high_flow,
    }

    final.append([duration, total_water_consumption, mean_flow_rate, max_flow_rate, min_flow_rate, flow_rate_changes, number_of_spikes, duration_of_high_flow])

  return final

def plot_data(features1, keys1, features2, keys2, features3, keys3):

    #print(keys1)
    #print(keys2)
    #print(keys3)

    line_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']


    # Convert features to DataFrame
    features_df_1 = pd.DataFrame(features1, columns=[
        'duration',
        'total_water_consumption',
        'mean_flow_rate',
        'max_flow_rate',
        'min_flow_rate',
        'flow_rate_changes',
        'number_of_spikes',
        'duration_of_high_flow'
    ])

    features_df_2 = pd.DataFrame(features2, columns=[
        'duration',
        'total_water_consumption',
        'mean_flow_rate',
        'max_flow_rate',
        'min_flow_rate',
        'flow_rate_changes',
        'number_of_spikes',
        'duration_of_high_flow'
    ])

    features_df_3 = pd.DataFrame(features3, columns=[
        'duration',
        'total_water_consumption',
        'mean_flow_rate',
        'max_flow_rate',
        'min_flow_rate',
        'flow_rate_changes',
        'number_of_spikes',
        'duration_of_high_flow'
    ])

    l = [2, 1, 0, 3, 4]
    features1 = features_df_1.iloc[:, l]
    features2 = features_df_2.iloc[:, l]
    features3 = features_df_3.iloc[:, l]

    # Standardize the features
    scaler = StandardScaler()
    scaled_features1 = scaler.fit_transform(features1)
    scaled_features2 = scaler.transform(features2)
    scaled_features3 = scaler.transform(features3)

    # Apply PCA to reduce to 2D for both dictionaries
    pca = PCA(n_components=2)
    pca_result1 = pca.fit_transform(scaled_features1)
    pca_result2 = pca.transform(scaled_features2)
    pca_result3 = pca.transform(scaled_features3)

    # Create a DataFrame with PCA results for both dictionaries
    pca_df1 = pd.DataFrame(pca_result1, columns=['PC1', 'PC2'])
    pca_df2 = pd.DataFrame(pca_result2, columns=['PC1', 'PC2'])
    pca_df3 = pd.DataFrame(pca_result3, columns=['PC1', 'PC2'])

    # Create scatter plot traces for both dictionaries
    scatter_trace1 = go.Scatter(
        x=pca_df1['PC1'],
        y=pca_df1['PC2'],
        mode='markers',
        marker=dict(size=10, color='blue'),
        name='Dictionary 1',
        text=[f'Index: {keys1[i]}<br>({x:.2f}, {y:.2f})' for i, (x, y) in enumerate(zip(pca_df1['PC1'], pca_df1['PC2']))],
        hoverinfo='text'
    )

    scatter_trace2 = go.Scatter(
        x=pca_df2['PC1'],
        y=pca_df2['PC2'],
        mode='markers',
        marker=dict(size=10, color='red'),
        name='Dictionary 2',
        text=[f'Index: {keys2[i]}<br>({x:.2f}, {y:.2f})' for i, (x, y) in enumerate(zip(pca_df2['PC1'], pca_df2['PC2']))],
        hoverinfo='text'
    )

    scatter_trace3 = go.Scatter(
        x=pca_df3['PC1'],
        y=pca_df3['PC2'],
        mode='markers',
        marker=dict(size=10, color='green'),
        name='Dictionary 3',
        text=[f'Index: {keys3[i]}<br>({x:.2f}, {y:.2f})' for i, (x, y) in enumerate(zip(pca_df3['PC1'], pca_df3['PC2']))],
        hoverinfo='text'
    )

    def parabola(x, a, b, c):
      return a * x**2 + b * x + c

    # Extend the range of x values for the parabolic curves
    x_curve_extended = np.linspace(min(pca_df1['PC1'].min(), pca_df2['PC1'].min(), pca_df3['PC1'].min()),
                                   max(pca_df1['PC1'].max(), pca_df2['PC1'].max(), pca_df3['PC1'].max()), 100)

    # Fit parabolic curves to connect corresponding points
    parabola_traces = []
    for i in range(len(pca_df1)):
        x_data = [pca_df1['PC1'][i], pca_df2['PC1'][i], pca_df3['PC1'][i]]
        y_data = [pca_df1['PC2'][i], pca_df2['PC2'][i], pca_df3['PC2'][i]]
        popt, _ = curve_fit(parabola, x_data, y_data)
        y_curve_extended = parabola(x_curve_extended, *popt)
        parabola_trace = go.Scatter(
            x=x_curve_extended,
            y=y_curve_extended,
            mode='lines',
            line=dict(color=line_colors[i], width=1),
            name=keys1[i]+keys2[i]+keys3[i],  # Legend based on line_legends list
            showlegend=True,  # Show legend for each line
        )
        parabola_traces.append(parabola_trace)

    # Create layout
    layout = go.Layout(
        title='PCA calibration plot with Extended Parabolic Curves',
        xaxis=dict(title='Principal Component 1'),
        yaxis=dict(title='Principal Component 2'),
        hovermode='closest'
    )

    # Add all traces to the figure data
    fig = go.Figure(data=[scatter_trace1, scatter_trace2, scatter_trace3] + parabola_traces, layout=layout)

    # Display interactive plot
    fig.show()
```

```python
import matplotlib.pyplot as plt

def plot_data(features1, keys1, features2, keys2, features3, keys3):

    # Define the number of random colors you want
    num_random_colors = len(keys1)

    # Generate a list of random colors
    random_colors = ['#{:06x}'.format(random.randint(0, 256**3-1)) for _ in range(num_random_colors)]

    # Shuffle the list of random colors
    random.shuffle(random_colors)

    # Convert features to DataFrame
    features_df_1 = pd.DataFrame(features1, columns=[
        'duration',
        'total_water_consumption',
        'mean_flow_rate',
        'max_flow_rate',
        'min_flow_rate',
        'flow_rate_changes',
        'number_of_spikes',
        'duration_of_high_flow'
    ])

    features_df_2 = pd.DataFrame(features2, columns=[
        'duration',
        'total_water_consumption',
        'mean_flow_rate',
        'max_flow_rate',
        'min_flow_rate',
        'flow_rate_changes',
        'number_of_spikes',
        'duration_of_high_flow'
    ])

    features_df_3 = pd.DataFrame(features3, columns=[
        'duration',
        'total_water_consumption',
        'mean_flow_rate',
        'max_flow_rate',
        'min_flow_rate',
        'flow_rate_changes',
        'number_of_spikes',
        'duration_of_high_flow'
    ])

    l = [2, 1, 0, 3, 4]
    features1 = features_df_1.iloc[:, l]
    features2 = features_df_2.iloc[:, l]
    features3 = features_df_3.iloc[:, l]

    # Standardize the features
    scaler = StandardScaler()
    scaled_features1 = scaler.fit_transform(features1)
    scaled_features2 = scaler.transform(features2)
    scaled_features3 = scaler.transform(features3)

    # Apply PCA to reduce to 2D for both dictionaries
    pca = PCA(n_components=2)
    pca_result1 = pca.fit_transform(scaled_features1)
    pca_result2 = pca.transform(scaled_features2)
    pca_result3 = pca.transform(scaled_features3)

    # Create a DataFrame with PCA results for both dictionaries
    pca_df1 = pd.DataFrame(pca_result1, columns=['PC1', 'PC2'])
    pca_df2 = pd.DataFrame(pca_result2, columns=['PC1', 'PC2'])
    pca_df3 = pd.DataFrame(pca_result3, columns=['PC1', 'PC2'])

    # Create an empty list to store the traces
    all_traces = []


    for i in range(len(keys1)):
        color = random_colors[i - 1]
        # Extract the first three points from each scatter plot
        title_of_line = keys1[i] + keys2[i] + keys3[i]
        names_of_points = [keys1[i], keys2[i], keys3[i]]
        x_values = [pca_df1['PC1'].iloc[i], pca_df2['PC1'].iloc[i], pca_df3['PC1'].iloc[i]]
        y_values = [pca_df1['PC2'].iloc[i], pca_df2['PC2'].iloc[i], pca_df3['PC2'].iloc[i]]

        # Fit a parabola to the data
        coefficients = np.polyfit(x_values, y_values, 2)  # Fit a second-degree polynomial (a parabola)

        # Generate extended range of x values for the plot
        x_plot_extended = np.linspace(min(x_values) - 30, max(x_values) + 30, 100)

        # Calculate corresponding y values using the fitted polynomial
        y_plot_extended = np.polyval(coefficients, x_plot_extended)

        # Create trace for the parabola
        parabola_trace = go.Scatter(
            x=x_plot_extended,
            y=y_plot_extended,
            mode='lines',
            name=title_of_line,  # Unique name for each parabola
            line=dict(color=color)  # Random color
        )

        # Append the parabola trace to the list of all traces
        all_traces.append(parabola_trace)

        # Create trace for the data points
        data_trace = go.Scatter(
            x=x_values,
            y=y_values,
            mode='markers+text',  # Added text mode
            name=title_of_line,
            marker=dict(color=color, size=10),
            hovertext=names_of_points,  # Custom text labels
            textposition='top center',  # Position of the text relative to the markers
            hoverinfo='text'
        )

        # Append the data trace to the list of all traces
        all_traces.append(data_trace)


    # Define layout
    layout = go.Layout(
        title='Plot of Multiple Parabolas with Data Points',
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        hovermode='closest'
    )

    # Create figure object
    fig = go.Figure(data=all_traces, layout=layout)

    # Show the plot
    fig.show()
```
